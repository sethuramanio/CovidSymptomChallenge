```python
import os 
```

```python
for files in os.listdir('main_states_folder/month_wise_state_partition/'):
    if files!='.DS_Store':
#         print(files)
        df=pd.read_csv(f'main_states_folder/month_wise_state_partition/{files}')
        df['date']=pd.to_datetime(df['date'])
        df['month']=df['date'].dt.month
        
        collection=df.groupby('month')
        loc_of_doc=f'main_states_folder/state_wise_full_file/{files[:2]}/'
        print(loc_of_doc)
        os.makedirs(loc_of_doc)
        
        for name,group in collection: 
            

            loc_of_each_month=f'{loc_of_doc}month_{name}.csv'
            group.to_csv(loc_of_each_month, index=False)
        

        
```
